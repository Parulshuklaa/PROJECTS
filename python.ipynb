```python
def tilingProblem(n, dp):
    if n <= 2:
        return n
    
    if n in dp:
        return dp[n]
    
    dp[n] = tilingProblem(n-1, dp) + tilingProblem(n-2, dp)
    return dp[n]


dp = {}
n = int(input("enter the number: "))
print(tilingProblem(n, dp))
```

```python
def count_binary_strings(n, dp={}):
    if n in dp:
        return dp[n]
    
    if n == 1:
        return 2
    if n == 2:
        return 3
    
    dp[n] = count_binary_strings(n-1, dp) + count_binary_strings(n-2, dp)
    return dp[n]


n = 3
print(count_binary_strings(n))  
```
